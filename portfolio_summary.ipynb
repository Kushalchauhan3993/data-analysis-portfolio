```python
import pandas as pd

projects = {
    "Project": [
        "Supermarket Sales",
        "Student Performance",
        "Weather Analysis",
        "COVID Trends",
        "Stock Market"
    ],
    "Domain": [
        "Retail",
        "Education",
        "Climate",
        "Healthcare",
        "Finance"
    ],
    "Visualizations": [5,5,5,5,5]
}

summary = pd.DataFrame(projects)

summary
```

```python
summary.describe()
```

```python
summary.to_csv("portfolio_summary.csv", index=False)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(summary["Project"], summary["Visualizations"])
plt.title("Visualizations per Project")
plt.ylabel("Number of Charts")
plt.xticks(rotation=45)
plt.show()
```
